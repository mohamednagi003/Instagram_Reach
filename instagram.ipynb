```python
import pandas as pd
import numpy as np
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack
from sklearn.svm import LinearSVR
```

# cleaning the data set
<ol>
dropping the duplicates and Null-values
</ol>

```python
df.drop_duplicates(inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

# applying vectorizer:
<ol>
To converting the caption and hashtags columns into sparse matrix

```python
vector = TfidfVectorizer()
```

```python
Z = vector.fit_transform(df["Caption"])
```

```python
Z
```

```python
V = vector.fit_transform(df["Hashtags"])
```

# label the y to be the Impressions:
<ol>
(going to be predicted)

```python
y = df["Impressions"].values
```

# Now we do not need those columns anymore

```python
df.drop(columns=["Impressions"],inplace=True)
```

```python
df.drop(columns=["Caption","Hashtags"],inplace=True)
```

# applying the rest of the data set to variable A

```python
A = df.values
```

# horizontal stacking the Z,V and A variables

```python
new = hstack([Z,V,A])
```

# new contains sparse matrix of captions and hashtags + rest of the data set

```python
new
```

# applying new to X

```python
X = new
```

# trying the linear reagression algo:

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
pre = lr.predict(X_test)
```

# its score is 0.98 could be better!

```python
lr.score(X_test,y_test)
```
